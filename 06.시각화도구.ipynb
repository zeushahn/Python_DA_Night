### Matplotlib
: 파이썬의 표준 시각화 도구

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# 행번호 0을 삭제

df.drop(
    0,
    axis='rows',
    inplace=True
)
df.head()
```

```python
# Index 재정리
df.reset_index(
    inplace=True,
    drop=True
)
df.head()
```

```python
# NaN 정리
# 현재 NaN을 상위의 값으로 채움

df.fillna(
    method='ffill', # bfill : 하위값으로 채우기
    inplace=True
)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# 컬럼 내용
df.columns
```

```python
# 년도 컬럼의 데이터가 Object로 되어 있어 숫자로 변환
df['1970'].unique()
```

```python
# 1970의 전체 갯수
print(f"전체갯수 : {len(df['1970'])}")

# -인 갯수
print(f"-인 갯수 : {df[df['1970'] == '-'].shape[0]}")

# -이 아닌 갯수
print(f"-인 갯수 : {df[df['1970'] != '-'].shape[0]}")
```

---
#### 년도별 컬럼에 대해 Object를 숫자로 변환

```python
# - -> 0으로 변경
df.replace(
    '-',
    0,
    inplace=True
)
df.head()
```

```python
df.info()
```

----
#### 전출입 시각화

```python
# 서울에서 타지역으로 이동한 데이터만 추출
df_seoul = \
    df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]

df_seoul.head()
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop(
    '전출지별',
    axis='columns',
    inplace=True
)
df_seoul.head()
```

```python
# 전입지별 -> 전입지 변경
df_seoul.rename(
    {
        '전입지별':'전입지'
    },
    axis='columns',
    inplace=True
)
df_seoul.head()
```

```python
# 전입지를 index로 변경
df_seoul.set_index(
    '전입지',
    inplace=True
)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구의 선그래프로 시각화 
sr_one = df_seoul.loc['경기도',:]
sr_one.head()
```

```python
# 선그래프
plt.plot(
    sr_one
)
plt.show()
```
